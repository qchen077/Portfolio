```python
import pandas as pd
from census import Census

API_KEY = "94ce0700ada284b3db8a52fb54a8ff7e71c9ba12"
c = Census(API_KEY)

years = list(range(2010, 2024))

fields = (
    'B01003_001E',  # total population

    # race
    'B02001_002E',  # white alone
    'B02001_003E',  # black alone
    'B02001_004E',  # american indian / alaska native alone
    'B02001_005E',  # asian alone
    'B02001_006E',  # native hawaiian / pacific islander alone
    'B02001_007E',  # some other race alone
    'B02001_008E',  # two or more races

    # ethnicity (Hispanic / Latino)
    'B03002_002E',  # Hispanic or Latino
    'B03002_003E',  # White alone, not Hispanic
    'B03002_004E',  # Black alone, not Hispanic
)

dfs = []

for y in years:
    data = c.acs5.state_county_tract(
        fields=fields,
        state_fips='11',
        county_fips='001',
        tract='*',
        year=y
    )

    df = pd.DataFrame(data)
    df["year"] = y
    dfs.append(df)

population_df = pd.concat(dfs, ignore_index=True)

population_df["GEOID"] = (
    population_df["state"] +
    population_df["county"] +
    population_df["tract"]
)

population_df = population_df.rename(columns={
    'B01003_001E': 'total_population',
    'B02001_002E': 'white_alone',
    'B02001_003E': 'black_alone',
    'B02001_004E': 'ai_an_alone',
    'B02001_005E': 'asian_alone',
    'B02001_006E': 'nh_pi_alone',
    'B02001_007E': 'some_other_race_alone',
    'B02001_008E': 'two_or_more_races',
    'B03002_002E': 'hispanic_or_latino',
    'B03002_003E': 'white_non_hispanic',
    'B03002_004E': 'black_non_hispanic',
})

population_df.head()
```

```python
race_cols = [
    'white_alone',
    'black_alone',
    'ai_an_alone',
    'asian_alone',
    'nh_pi_alone',
    'some_other_race_alone',
    'two_or_more_races',
]

for col in race_cols:
    population_df[col + "_share"] = (
        population_df[col].astype(float) /
        population_df["total_population"].astype(float)
    )

population_df["hispanic_share"] = (
    population_df["hispanic_or_latino"].astype(float) /
    population_df["total_population"].astype(float)
)
```

```python
import geopandas as gpd

crashes_mpd = crashes.dropna(
    subset=["MPDLATITUDE", "MPDLONGITUDE"]
).copy()

crashes_mpd["geometry"] = gpd.points_from_xy(
    crashes_mpd["MPDLONGITUDE"],
    crashes_mpd["MPDLATITUDE"]
)

crashes_gdf = gpd.GeoDataFrame(
    crashes_mpd,
    geometry="geometry",
    crs="EPSG:4326"
)
```

```python
crashes_gdf.crs
```

```python
# Spatial join 1: crashes - census tract
crash_tract = gpd.sjoin(
    crashes_gdf,
    tract,
    how='left',
    predicate='within'
)
```

```python
crash_count = crash_tract.groupby("GEOID").size().reset_index(name="crash_count")

crash_summary = (
    crash_tract.groupby("GEOID")
    .agg(
        crash_count = ("CRIMEID", "count"),
        speeding_count = ("SPEEDING_INVOLVED", "sum"),
        total_vehicles = ("TOTAL_VEHICLES", "sum"),
        fatal_driver = ("FATAL_DRIVER", "sum"),
        fatal_pedestrian = ("TOTAL_PEDESTRIANS", "sum"),
        fata_bicyclist = ("FATAL_BICYCLIST", "sum"),
        major_inj_driver = ("MAJORINJURIES_DRIVER", "sum"),
        major_inj_pred = ("MAJORINJURIES_PEDESTRIAN", "sum"),
        major_inj_bike = ("MAJORINJURIES_BICYCLIST", "sum")
        
    ).reset_index()
)
```

```python
# Step 3: roadwayBI - convert to GeoDataFrame, then spatial join tract

roadwayBI_gdf = gpd.GeoDataFrame(
    roadwayBI.dropna(subset=["X", "Y"]).copy(),
    geometry=gpd.points_from_xy(roadwayBI["X"], roadwayBI["Y"]),
    crs="EPSG:3857"
).to_crs(tract.crs)

bi_tract = gpd.sjoin(roadwayBI_gdf, tract, how="left", predicate="within")

intersection_summary = (
    bi_tract.groupby("GEOID")
    .size()
    .reset_index(name="intersection_count")
)
```

```python
# Step 4: traffic - LATITUDE/ LONGTITUDE

traffic_gdf = gpd.GeoDataFrame(
    traffic.dropna(subset=["LATITUDE", "LONGITUDE"]).copy(),
    geometry = gpd.points_from_xy(
        traffic["LONGITUDE"],
        traffic["LATITUDE"]
    ),
    crs="EPSG:4326"
).to_crs(tract.crs)

traffic_tract = gpd.sjoin(traffic_gdf, tract, how="left", predicate="within")

signal_summary = (
    traffic_tract.groupby("GEOID")
    .agg(
        signal_count = ("OBJECTID", "count"),
        aps_count = ("APS", lambda s: (s == "Yes").sum() if s.dtype == object else 0),
        hawk_count = ("HAWK", lambda s: (s == "Yes").sum() if s.dtype == object else 0)
    )
)
```
